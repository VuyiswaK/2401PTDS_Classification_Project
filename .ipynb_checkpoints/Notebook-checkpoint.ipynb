## Import libraries and read in the data

```python
# Import necessary libraries

#miscallenous
import pandas as pd 
import numpy as np


#NLP
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, TreebankWordTokenizer

# plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import re

# set plot style
sns.set_theme()
```

```python
# Read the train dataset
train = pd.read_csv('https://raw.githubusercontent.com/Jana-Liebenberg/2401PTDS_Classification_Project/refs/heads/main/Data/processed/train.csv')
train.head()
```

```python
# Read the test dataset
test = pd.read_csv('https://raw.githubusercontent.com/Jana-Liebenberg/2401PTDS_Classification_Project/refs/heads/main/Data/processed/test.csv')
test.head()
```

## Data Cleaning

We remove all punctuation, symbols, spaces, non-characters, and quotes. This can be done with regex

```python
def clean(text):
    # Remove commas
    text = re.sub(r',', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    # Remove full stops
    text = re.sub(r'\.', '', text)
    # Remove single quotes and double quotes
    text = re.sub(r"['\"]", '', text)
    # Remove other non-word characters
    text = re.sub(r'\W', ' ', text)
    
    #tokenize
    tokeniser = TreebankWordTokenizer()
    text_token = tokeniser.tokenize(text)
    
    tokens_without_sw = [word for word in text_token if not word in stop_words]
    
    return text #tokens_without_sw 
```

```python
# Tokenise the text using the TreebankWordTokenizer and remove stop words

# Download stopwords
#nltk.download('stopwords')
#nltk.download('punkt')

stop_words = set(stopwords.words('english'))

train_copy = train.copy()
for col in train_copy.columns:
    if col != 'category':
        train_copy[col] = train_copy[col].apply(clean)
#train_copy['url'] = train_copy['url'].apply(clean)
```

```python
train_copy
```

### EDA

```python
grouped = train.groupby(['category']).count().reset_index().sort_values(by = 'headlines', ascending = False)
plt.bar(grouped.category, grouped.headlines)
```

MODELLING

```python
X_train = train_copy.drop('category',axis = 1)['url']
y_train = train_copy['category']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from sklearn.pipeline import Pipeline
lr = Pipeline([('tfidf', TfidfVectorizer()),
               ('clf', LogisticRegression(max_iter = 1000)),
              ])

# Train the logistic regression model on the training set

lr.fit(X_train,y_train)
```

```python
# Make predictions on the test set
X_test = test['description'].apply(clean)
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy of the model
y_test = test['category']
print(f"Accuracy is: {accuracy_score(y_pred,y_test)}")
```
